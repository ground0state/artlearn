```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import MinMaxScaler

from artlearn import ART2

sns.set()
```

```python
def prepare_data():
    X, y = make_blobs(n_samples=300, centers=6, n_features=2, cluster_std=0.6, random_state=0)
    scaler = MinMaxScaler()
    X = scaler.fit_transform(X)
    X_ = X.copy()
    X_ = np.hstack([X, 1.0 - X_])
    return X_, y
```

```python
# prepare data
X, y = prepare_data()
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="Set2")
plt.show()
```

```python
# learning
clf = ART2(max_iter=100, max_class=100, a=0.01, b=0.01, c=0.1, d=0.9, theta=0.0, rho=0.99)
clf.fit(X)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=clf.labels_, palette="Set2")
plt.show()
```
